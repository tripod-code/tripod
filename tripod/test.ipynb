```python
from dustpy import hdf5writer
import matplotlib.pyplot as plt
import dustpy.constants as c

wrtr = hdf5writer()
```

```python
data = wrtr.read.all()
```

```python
plt.figure(figsize=(10, 6))
for it in range(len(data.t)):
    plt.plot(data.grid.r[it,:]/c.au, data.gas.Sigma[it,:], label='Gas Surface Density')
    plt.plot(data.grid.r[it,:]/c.au, data.gas.components.default.Sigma[it,:], "--",label='Dust Surface Density')
    plt.xscale('log')
    plt.yscale('log')
    plt.ylim(1e-5, 1e2)
    plt.title('Surface Density at t = {:.2f} yr'.format(data.t[it]/c.year))
    plt.ylabel('Surface Density (g/cm)')
    plt.xlabel('Radius (AU)')
```
